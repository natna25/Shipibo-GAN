```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import image

import torch
import os

from PIL import Image
```

```python



# load image as pixel array
data = image.imread('assets/test.jpg')
# summarize shape of the pixel array
print(data.dtype)
print(data.shape)
# display the array of pixels as an image
plt.imshow(data)
plt.show()
```

```python
#example resize of the image

img = Image.open('assets/test.jpg')

print(img.size)
plt.imshow(img)
#resizes image while preserving aspect ratio
img.thumbnail((256,256))
print(img.size)
plt.imshow(img)

```

```python
img = Image.open('assets/test.jpg')

print(img.size)
plt.imshow(img)
#resizes image while preserving aspect ratio
resized = img.resize((256,256))
print(resized.size)
plt.imshow(resized)
```
